```python
x = df_yards["Year"]
y = df_yards["MeanYards"]

m, b = np.polyfit(x, y, 1)  #I'm trying to find a line of best fit for how the data changed over time so I use the linear formula y = mx + b


plt.scatter(x, y)
plt.plot(x, m*x + b)
plt.title('Average yards per year')
plt.xlabel('Year')
plt.ylabel('Yards')
plt.show()
```
